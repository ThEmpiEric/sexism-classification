```python
def numIslands(grid: List[List[str]]) -> int:
    m, n = len(grid), len(grid[0])
    marked = [ [False] * n  ] * m  

    def dfs(i,j):
        if i >= m or i < 0 or  j >= n or j < 0 or grid[i][j] != '1': 
            return
        else:
            grid[i][j] = '0'
            posiciones = [ [1,0], [0,1], [-1, 0], [0, -1] ] 
            for pos in posiciones: 
                new_pos = [ a + b for a, b in zip( [i, j], pos) ] 
                dfs(new_pos[0],new_pos[1]) 

    number_islands = 0 
    for i in range(m): 
        for j in range(n): 
            if grid[i][j] == '1':
                number_islands += 1 
                dfs(i,j)
    
    return number_islands 



```
